```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import Thermobar as pt
import pyOpam as O

import matplotlib as mpl
mpl.rcParams['axes.titlesize'] = 16
mpl.rcParams['axes.labelsize'] = 14
mpl.rcParams['xtick.labelsize'] = 12
mpl.rcParams['ytick.labelsize'] = 12
O.__version__
```

```python
Data = pd.read_excel('NewOpamExperiments.xlsx')
OPAM = Data.copy() # only OPAM saturated experimnts
No_OPAM = Data.copy() # all experiments

A = (Data['Ol'] == 'Yes') & (Data['Plag'] == 'Yes') & (Data['Cpx'] == 'Yes')
OPAM = OPAM[A]
OPAM['MnO_Liq'][np.isnan(OPAM['MnO_Liq'])] = 0.0
OPAM['Cr2O3_Liq'][np.isnan(OPAM['Cr2O3_Liq'])] = 0.0
OPAM['P2O5_Liq'][np.isnan(OPAM['P2O5_Liq'])] = 0.0
OPAM = OPAM.reset_index(drop = True)
```

```python
OPAM['Sample_ID_Liq'] = np.linspace(0,len(OPAM['SiO2_Liq'])-1, len(OPAM['SiO2_Liq']))
Results_Yang = O.calc_liq_press_temp(liq_comps = OPAM, 
                                     equationP = "P_Yang1996", 
                                     equationT = "T_Yang1996")
Results_Voigt = O.calc_liq_press_temp(liq_comps = OPAM, 
                                      equationP = "P_Voigt2017", 
                                      equationT = "T_Voigt2017", 
                                      fo2 = 10**(OPAM['logfo2']))
Results_Herzberg = O.calc_liq_press(liq_comps = OPAM,
                                   equationP = "P_Herzberg2004")
Results_Villiger = O.calc_liq_press(liq_comps = OPAM,
                                   equationP = "P_Villiger2007")
```

```python
f, a = plt.subplots(2,2, figsize = (10,8))
a[0][0].plot([0, 10], [0,10], '--k')
a[0][0].plot(Results_Yang['P_kbar'], Results_Yang['P_kbar_calc'], 'or', ms = 4)
a[0][0].plot(Results_Yang['P_kbar'][Results_Yang['Pf'] > 0.8], 
             Results_Yang['P_kbar_calc'][Results_Yang['Pf'] > 0.8], 'db', ms = 8)

a[0][1].plot([1050, 1250], [1050, 1250], '--k')
a[0][1].plot(Results_Yang['T_K'] - 273.15, Results_Yang['T_C_calc'], 'or', ms = 4)
a[0][1].plot(Results_Yang['T_K'][Results_Yang['Pf'] > 0.8] - 273.15, 
             Results_Yang['T_C_calc'][Results_Yang['Pf'] > 0.8], 'db', ms = 8)


a[1][0].plot([0, 10], [0,10], '--k')
a[1][0].plot(Results_Voigt['P_kbar'], Results_Voigt['P_kbar_calc'], 'or', ms = 4)
a[1][0].plot(Results_Voigt['P_kbar'][Results_Voigt['Pf'] > 0.8], 
             Results_Voigt['P_kbar_calc'][Results_Voigt['Pf'] > 0.8], 'db', ms = 8)

a[1][1].plot([1050, 1250], [1050, 1250], '--k')
a[1][1].plot(Results_Voigt['T_K'] - 273.15, Results_Voigt['T_C_calc'], 'or', ms = 4)
a[1][1].plot(Results_Voigt['T_K'][Results_Voigt['Pf'] > 0.8] - 273.15, 
             Results_Voigt['T_C_calc'][Results_Voigt['Pf'] > 0.8], 'db', ms = 8)
```

```python
Colls = [col for col in Data.columns if '_Liq' in col]

Liqs = pd.DataFrame()
for C in Colls:
    Liqs[C] = OPAM[C].copy()

Liqs['Sample_ID_Liq'] = np.linspace(0, len(Liqs['SiO2_Liq'])-1, len(Liqs['SiO2_Liq']))
#Liqs['Cr2O3_Liq'][np.isnan(Liqs['Cr2O3_Liq'])] = 0.05

No_Liqs = pd.DataFrame()
for C in Colls:
    No_Liqs[C] = No_OPAM[C].copy()

No_Liqs['Sample_ID_Liq'] = np.linspace(0, len(No_Liqs['SiO2_Liq'])-1, len(No_Liqs['SiO2_Liq']))

Arc_Liqs = pd.DataFrame()
for C in Colls:
    Arc_Liqs[C] = Arc[C].copy()

Arc_Liqs['Sample_ID_Liq'] = np.linspace(0, len(Arc_Liqs['SiO2_Liq'])-1, len(Arc_Liqs['SiO2_Liq']))
```

```python
Res_V_all = Opam.pyOpam(liq_comps = No_Liqs, equationP = "Voigt2017", fo2 = "QFM")
Res_V_all_fin = Res_V_all.copy()
Res_V_all_fin =Res_V_all_fin.fillna(0)
V_all_results = pt.calculate_R2(No_OPAM['P_kbar'], Res_V_all['P_kbar_calc'])
V_all_results_test = pt.calculate_R2(No_OPAM['P_kbar'][Res_V_all['Pf'] > 0.8], 
                                     Res_V_all['P_kbar_calc'][Res_V_all['Pf'] > 0.8])
V_all_RMSE = np.sqrt((1/len(No_OPAM['SiO2_Liq']))*np.sum((Res_V_all['P_kbar_calc'] - No_OPAM['P_kbar'])**2))
V_all_RMSE_test = np.sqrt((1/len(No_OPAM['SiO2_Liq'][Res_V_all['Pf'] > 0.8]))*
                     np.sum((Res_V_all['P_kbar_calc'][Res_V_all['Pf'] > 0.8]
                             - No_OPAM['P_kbar'][Res_V_all['Pf'] > 0.8])**2))
V_all_MAE = (1/len(No_OPAM['SiO2_Liq']))*np.sum(Res_V_all['P_kbar_calc'] - No_OPAM['P_kbar'])
V_all_MAE_test = (1/len(No_OPAM['SiO2_Liq'][Res_V_all['Pf'] > 0.8]))*np.sum(Res_V_all['P_kbar_calc'][Res_V_all['Pf'] > 0.8]
                                                                      - No_OPAM['P_kbar'][Res_V_all['Pf'] > 0.8])

Res_V_OPAM = Opam.pyOpam(liq_comps = Liqs, equationP = "Voigt2017", fo2 = "QFM")
Res_V_OPAM_fin = Res_V_OPAM.copy()
Res_V_OPAM_fin =Res_V_OPAM_fin.fillna(0)
V_OPAM_results = pt.calculate_R2(OPAM['P_kbar'], Res_V_OPAM['P_kbar_calc'])
V_OPAM_results_test = pt.calculate_R2(OPAM['P_kbar'][Res_V_OPAM['Pf'] > 0.8], 
                                      Res_V_OPAM['P_kbar_calc'][Res_V_OPAM['Pf'] > 0.8])
V_OPAM_RMSE = np.sqrt((1/len(OPAM['SiO2_Liq']))*np.sum((Res_V_OPAM['P_kbar_calc'] - OPAM['P_kbar'])**2))
V_OPAM_RMSE_test = np.sqrt((1/len(OPAM['SiO2_Liq'][Res_V_OPAM['Pf'] > 0.8]))*
                     np.sum((Res_V_OPAM['P_kbar_calc'][Res_V_OPAM['Pf'] > 0.8]
                             - OPAM['P_kbar'][Res_V_OPAM['Pf'] > 0.8])**2))
V_OPAM_MAE = (1/len(OPAM['SiO2_Liq']))*np.sum(Res_V_OPAM['P_kbar_calc'] - OPAM['P_kbar'])
V_OPAM_MAE_test = (1/len(OPAM['SiO2_Liq'][Res_V_OPAM['Pf'] > 0.8]))*np.sum(Res_V_OPAM['P_kbar_calc'][Res_V_OPAM['Pf'] > 0.8]
                                                                      - OPAM['P_kbar'][Res_V_OPAM['Pf'] > 0.8])

Res_V_Arc = Opam.pyOpam(liq_comps = Arc_Liqs, equationP = "Voigt2017", fo2 = "QFM")
Res_V_Arc_fin = Res_V_Arc.copy()
Res_V_Arc_fin =Res_V_Arc_fin.fillna(0)
V_Arc_results = pt.calculate_R2(Arc['P_kbar'], Res_V_Arc['P_kbar_calc'])
V_Arc_results_test = pt.calculate_R2(Arc['P_kbar'][Res_V_Arc['Pf'] > 0.8], 
                                      Res_V_Arc['P_kbar_calc'][Res_V_Arc['Pf'] > 0.8])
V_Arc_RMSE = np.sqrt((1/len(Arc['SiO2_Liq']))*np.sum((Res_V_Arc['P_kbar_calc'] - Arc['P_kbar'])**2))
V_Arc_RMSE_test = np.sqrt((1/len(Arc['SiO2_Liq'][Res_V_Arc['Pf'] > 0.8]))*
                     np.sum((Res_V_Arc['P_kbar_calc'][Res_V_Arc['Pf'] > 0.8]
                             - Arc['P_kbar'][Res_V_Arc['Pf'] > 0.8])**2))
V_Arc_MAE = (1/len(Arc['SiO2_Liq']))*np.sum(Res_V_Arc['P_kbar_calc'] - Arc['P_kbar'])
V_Arc_MAE_test = (1/len(Arc['SiO2_Liq'][Res_V_Arc['Pf'] > 0.8]))*np.sum(Res_V_Arc['P_kbar_calc'][Res_V_Arc['Pf'] > 0.8]
                                                                      - Arc['P_kbar'][Res_V_Arc['Pf'] > 0.8])

Res_H = Opam.Herzberg2004(Liqs)
Res_H_fin = Res_H.copy()
Res_H_fin = Res_H_fin.fillna(0)
H_results = pt.calculate_R2(OPAM['P_kbar'], Res_H_fin['P_kbar_calc'])
H_MAE = (1/len(OPAM['SiO2_Liq']))*np.sum(Res_H_fin['P_kbar_calc'] - OPAM['P_kbar'])

Res_H_all = Opam.Herzberg2004(No_Liqs)
Res_H_all_fin = Res_H_all.copy()
Res_H_all_fin = Res_H_all_fin.fillna(0)
H_all_results = pt.calculate_R2(No_OPAM['P_kbar'], Res_H_all_fin['P_kbar_calc'])
H_all_MAE = (1/len(No_OPAM['SiO2_Liq']))*np.sum(Res_H_all_fin['P_kbar_calc'] - No_OPAM['P_kbar'])

Res_H_Arc = Opam.Herzberg2004(Arc_Liqs)
Res_H_Arc_fin = Res_H_Arc.copy()
Res_H_Arc_fin = Res_H_Arc_fin.fillna(0)
H_Arc_results = pt.calculate_R2(Arc['P_kbar'], Res_H_Arc_fin['P_kbar_calc'])
H_Arc_MAE = (1/len(Arc['SiO2_Liq']))*np.sum(Res_H_Arc_fin['P_kbar_calc'] - Arc['P_kbar'])

Res_Y_all = Opam.pyOpam(liq_comps = No_Liqs, equationP = "Yang1996")
Res_Y_all_fin = Res_Y_all.copy()
Res_Y_all_fin =Res_Y_all_fin.fillna(0)
Y_all_results = pt.calculate_R2(No_OPAM['P_kbar'], Res_Y_all['P_kbar_calc'])
Y_all_results_test = pt.calculate_R2(No_OPAM['P_kbar'][Res_Y_all['Pf'] > 0.8], 
                                     Res_Y_all['P_kbar_calc'][Res_Y_all['Pf'] > 0.8])
Y_all_RMSE = np.sqrt((1/len(No_OPAM['SiO2_Liq']))*np.sum((Res_Y_all['P_kbar_calc'] - No_OPAM['P_kbar'])**2))
Y_all_RMSE_test = np.sqrt((1/len(No_OPAM['SiO2_Liq'][Res_Y_all['Pf'] > 0.8]))*
                     np.sum((Res_Y_all['P_kbar_calc'][Res_Y_all['Pf'] > 0.8]
                             - No_OPAM['P_kbar'][Res_Y_all['Pf'] > 0.8])**2))
Y_all_MAE = (1/len(No_OPAM['SiO2_Liq']))*np.sum(Res_Y_all['P_kbar_calc'] - No_OPAM['P_kbar'])
Y_all_MAE_test = (1/len(No_OPAM['SiO2_Liq'][Res_Y_all['Pf'] > 0.8]))*np.sum(Res_Y_all['P_kbar_calc'][Res_Y_all['Pf'] > 0.8]
                                                                      - No_OPAM['P_kbar'][Res_Y_all['Pf'] > 0.8])

Res_Y_OPAM = Opam.pyOpam(liq_comps = Liqs, equationP = "Yang1996")
Res_Y_OPAM_fin = Res_Y_OPAM.copy()
Res_Y_OPAM_fin = Res_Y_OPAM_fin.fillna(0)
Y_OPAM_results = pt.calculate_R2(OPAM['P_kbar'], Res_Y_OPAM['P_kbar_calc'])
Y_OPAM_results_test = pt.calculate_R2(OPAM['P_kbar'][Res_Y_OPAM['Pf'] > 0.8], 
                                      Res_Y_OPAM['P_kbar_calc'][Res_Y_OPAM['Pf'] > 0.8])
Y_OPAM_RMSE = np.sqrt((1/len(OPAM['SiO2_Liq']))*np.sum((Res_Y_OPAM['P_kbar_calc'] - OPAM['P_kbar'])**2))
Y_OPAM_RMSE_test = np.sqrt((1/len(OPAM['SiO2_Liq'][Res_Y_OPAM['Pf'] > 0.8]))*
                     np.sum((Res_Y_OPAM['P_kbar_calc'][Res_Y_OPAM['Pf'] > 0.8]
                             - OPAM['P_kbar'][Res_Y_OPAM['Pf'] > 0.8])**2))
Y_OPAM_MAE = (1/len(OPAM['SiO2_Liq']))*np.sum(Res_Y_OPAM['P_kbar_calc'] - OPAM['P_kbar'])
Y_OPAM_MAE_test = (1/len(OPAM['SiO2_Liq'][Res_Y_OPAM['Pf'] > 0.8]))*np.sum(Res_Y_OPAM['P_kbar_calc'][Res_Y_OPAM['Pf'] > 0.8]
                                                                      - OPAM['P_kbar'][Res_Y_OPAM['Pf'] > 0.8])

Res_Y_Arc = Opam.pyOpam(liq_comps = Arc_Liqs, equationP = "Yang1996")
Res_Y_Arc_fin = Res_Y_Arc.copy()
Res_Y_Arc_fin = Res_Y_Arc_fin.fillna(0)
Y_Arc_results = pt.calculate_R2(Arc['P_kbar'], Res_Y_Arc['P_kbar_calc'])
Y_Arc_results_test = pt.calculate_R2(Arc['P_kbar'][Res_Y_Arc['Pf'] > 0.8], 
                                      Res_Y_Arc['P_kbar_calc'][Res_Y_Arc['Pf'] > 0.8])
Y_Arc_RMSE = np.sqrt((1/len(Arc['SiO2_Liq']))*np.sum((Res_Y_Arc['P_kbar_calc'] - Arc['P_kbar'])**2))
Y_Arc_RMSE_test = np.sqrt((1/len(Arc['SiO2_Liq'][Res_Y_Arc['Pf'] > 0.8]))*
                     np.sum((Res_Y_Arc['P_kbar_calc'][Res_Y_Arc['Pf'] > 0.8]
                             - Arc['P_kbar'][Res_Y_Arc['Pf'] > 0.8])**2))
Y_Arc_MAE = (1/len(Arc['SiO2_Liq']))*np.sum(Res_Y_Arc['P_kbar_calc'] - Arc['P_kbar'])
Y_Arc_MAE_test = (1/len(Arc['SiO2_Liq'][Res_Y_Arc['Pf'] > 0.8]))*np.sum(Res_Y_Arc['P_kbar_calc'][Res_Y_Arc['Pf'] > 0.8]
                                                                      - Arc['P_kbar'][Res_Y_Arc['Pf'] > 0.8])

Results = {}
Results['Y_OPAM'] = {'Results': Res_Y_OPAM_fin, 'Stats': Y_OPAM_results, 'Stats_test': Y_OPAM_results_test,
                    'MAE': Y_OPAM_MAE, 'MAE_test': Y_OPAM_MAE_test, 'P_kbar_true': OPAM['P_kbar']}

Results['V_OPAM'] = {'Results': Res_V_OPAM_fin, 'Stats': V_OPAM_results, 'Stats_test': V_OPAM_results_test,
                    'MAE': V_OPAM_MAE, 'MAE_test': V_OPAM_MAE_test, 'P_kbar_true': OPAM['P_kbar']}

Results['H_OPAM'] = {'Results': Res_H_fin, 'Stats': H_results, 'Stats_test': np.nan,
                    'MAE': H_MAE, 'MAE_test': np.nan, 'P_kbar_true': OPAM['P_kbar']}

Results['Y_all'] = {'Results': Res_Y_all_fin, 'Stats': Y_all_results, 'Stats_test': Y_all_results_test,
                    'MAE': Y_all_MAE, 'MAE_test': Y_all_MAE_test, 'P_kbar_true': Data['P_kbar']}

Results['V_all'] = {'Results': Res_V_all_fin, 'Stats': V_all_results, 'Stats_test': V_all_results_test,
                    'MAE': V_all_MAE, 'MAE_test': V_all_MAE_test, 'P_kbar_true': Data['P_kbar']}

Results['H_all'] = {'Results': Res_H_all_fin, 'Stats': H_all_results, 'Stats_test': np.nan,
                    'MAE': H_all_MAE, 'MAE_test': np.nan, 'P_kbar_true': Data['P_kbar']}

Results['Y_Arc'] = {'Results': Res_Y_Arc_fin, 'Stats': Y_Arc_results, 'Stats_test': Y_Arc_results_test,
                    'MAE': Y_Arc_MAE, 'MAE_test': Y_Arc_MAE_test, 'P_kbar_true': Arc['P_kbar']}

Results['V_Arc'] = {'Results': Res_V_Arc_fin, 'Stats': V_Arc_results, 'Stats_test': V_Arc_results_test,
                    'MAE': V_Arc_MAE, 'MAE_test': V_Arc_MAE_test, 'P_kbar_true': Arc['P_kbar']}

Results['H_Arc'] = {'Results': Res_H_Arc_fin, 'Stats': H_Arc_results, 'Stats_test': np.nan,
                    'MAE': H_Arc_MAE, 'MAE_test': np.nan, 'P_kbar_true': Arc['P_kbar']}
```

```python
Elements = np.array(['SiO2_Liq', 'TiO2_Liq', 'Al2O3_Liq', 'FeOt_Liq', 'MnO_Liq', 
                                     'MgO_Liq', 'CaO_Liq', 'Na2O_Liq', 'K2O_Liq'])
Base = MG_fin[MG_fin['Pf'] > 0.8][Elements].loc[0]
Ol = Hosts[Elements][Hosts['host'] == 'olv']
Ol = Ol.replace('-', 0.0)
Ol_add = Ol.mean()

Cp = Hosts[Elements][Hosts['host'] == 'cpx']
Cp = Cp.replace('-', 0.0)
Cp_add = Cp.mean()

Pl = Hosts[Elements][Hosts['host'] == 'plg']
Pl = Pl.replace('-', 0.0)
Pl_add = Pl.mean()

Gab_add = Ol_add*0.05 + Pl_add*0.5 + Cp_add*0.45

x = np.linspace(-0.1,0.3,300)
Add_ol = pd.DataFrame(data = np.zeros((len(x),11)), columns = ['Sample_ID_Liq', 'Ol_added'] + list(Base.keys()))
Add_pl = pd.DataFrame(data = np.zeros((len(x),11)), columns = ['Sample_ID_Liq', 'Pl_added'] + list(Base.keys()))
Add_cp = pd.DataFrame(data = np.zeros((len(x),11)), columns = ['Sample_ID_Liq', 'Cp_added'] + list(Base.keys()))
Add_gab = pd.DataFrame(data = np.zeros((len(x),11)), columns = ['Sample_ID_Liq', 'Gab_added'] + list(Base.keys()))

for i in range(len(Add_ol['SiO2_Liq'])):
    Add_ol.loc[i] = Base*(1-x[i]) + Ol_add*x[i]
    Add_ol['Sample_ID_Liq'].loc[i] = i
    
    Add_pl.loc[i] = Base*(1-x[i]) + Pl_add*x[i]
    Add_pl['Sample_ID_Liq'].loc[i] = i
    
    Add_cp.loc[i] = Base*(1-x[i]) + Cp_add*x[i]
    Add_cp['Sample_ID_Liq'].loc[i] = i
    
    Add_gab.loc[i] = Base*(1-x[i]) + Gab_add*x[i]
    Add_gab['Sample_ID_Liq'].loc[i] = i
    
    
Add_ol['Ol_added'] = x
Add_pl['Pl_added'] = x
Add_cp['Cp_added'] = x
Add_gab['Gab_added'] = x
```
